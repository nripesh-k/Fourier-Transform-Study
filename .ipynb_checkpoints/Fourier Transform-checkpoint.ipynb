```python
def plotSignal(signal,reflect = 0):
    %matplotlib inline
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(3)
    plt.plot([s[0] for s in signal],[s[1] for s in signal],c='m')
    if reflect:
        plt.plot([-s[0] for s in signal],[s[1] for s in signal],c='m') # fourier transform even function
    plt.show()
```

```python
plotSignal(signal)
```

```python
def add(a,b):
    x = a[0]+b[0]
    y = a[1]+b[1]
    return [x,y]

def sub(a,b):
    x = a[0]-b[0]
    y = a[1]-b[1]
    return [x,y]

def mult(a,b):
    x = a[0]*b[0]-a[1]*b[1]
    y = a[0]*b[1]+a[1]*b[0]
    return [x,y]

def magnitude(a,b):
    return a*a+b*b
```

```python
def low_pass(sigX,sig,maxm):
    for x,s in zip(sigX,sig):
        if x['f']>maxm:
            x['r'],x['i'],x['amp'],x['phase'] = 0,0,0,0
            s[1]=0
```

```python
def dtft(x):
    Complex = []
    X = []
    N = len(x)
    k = 0
    while k<N:
        sum = [0,0]
        try:     # for complex signal i.e, <t,(x,iy)>
            x[1][1][0]
            for n in range(N):
                phi = (m.pi * 2 * k * n) / N        # e^-(2pi*i*k*n/N)  =>  clockwise winding of the function
                c = [m.cos(phi), -m.sin(phi)]       # e^(x*i) = cos(x) - sin(x)
                sum = add(sum,mult([x[n][1][0],x[n][1][1]],c)) # x[n] => [x[n][1],0]
        except:  # for normal signal i.e, <t,x>
            for n in range(N):
                phi = (m.pi * 2 * k * n) / N
                c = [m.cos(phi), -m.sin(phi)]
                sum = add(sum,mult([x[n][1],0],c)) 
        sum[0] = 2*sum[0] / N
        sum[1] = 2*sum[1] / N

        freq = k
        amp = m.sqrt(sum[0] * sum[0] + sum[1] * sum[1])
        phase = m.atan2(sum[1], sum[0])
        X.append({ 'r': sum[0], 'i': sum[1], 'f': freq, 'amp': amp, 'phase': phase })
        Complex.append([freq,magnitude(sum[0],sum[1])])
        k+=1
    low_pass(X,Complex,N/2)
    return X,Complex;
```

```python
ftX,ft = dtft(signal)
```

```python
plotSignal(ft)
```

```python
def i_dtft(x):
    Complex = []
    X = []
    N = len(x)
    k = 0
    while k<N:
        sum = [0,0]
        for n in range(N):
            phi = (m.pi * 2 * k * n) / N
            c = [m.cos(phi), m.sin(phi)]
            sum = add(sum,mult([x[n]['r'],x[n]['i']],c)) # x[n] => [x[n][1],0]
        freq = k
        amp = m.sqrt(sum[0] * sum[0] + sum[1] * sum[1])
        phase = m.atan2(sum[1], sum[0])
        X.append({ 'r': sum[0], 'i': sum[1], 'f': freq, 'amp': amp, 'phase': phase })
        Complex.append([freq,magnitude(sum[0],sum[1])])
        k+=1.0    # k+=1/scale
    return X,Complex;
```

```python
def plotInverse(x):
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(3)
    plt.plot([s['f'] for s in x],[s['r'] for s in x],c='g')   # 2*s['f'] => dtft N/2
#     plt.plot([s['f'] for s in x],[s['i'] for s in x],c='g')
    plt.show()
```

```python
def processSignal(signal, reflect = 0):
    %matplotlib inline
    plotSignal(signal)
    Xft,ft = dtft(signal)
#     plotSignal(ft,reflect)
    plotSignal(ft[:len(ft)//2],reflect)
    iXft,ift = i_dtft(Xft)
    plotInverse(iXft)
```

```python
def overlayPlot(signals):
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(3)
    for signal in signals:
        if signal[0]:   # noraml signal
            plt.plot([s[0] for s in signal[1]],[s[1] for s in signal[1]],c='c')
        else:           # complex signal
            plt.plot([x['f'] for x in signal[1]],[x['r'] for x in signal[1]],c='m')
    plt.show()
```

```python
def lowPassProcess(signal,maxm):
    ftX,ft = dtft(signal)
    plotSignal(ft)
    low_pass(ftX,ft,maxm)
    plotSignal(ft)
    iftX,ift = i_dtft(ftX)
    overlayPlot([[1,signal],[0,iftX]])
```

```python
def lowEnergy(sigX,sig,threshold):
    for x,s in zip(sigX,sig):
        if s[1]<threshold:
            x['r'],x['i'],x['amp'],x['phase'] = 0,0,0,0
            s[1]=0
```

```python
def lowEnergyProcess(signal,threshold):
    ftX,ft = dtft(signal)
    plotSignal(ft)
    lowEnergy(ftX,ft,threshold)
    plotSignal(ft)
    iftX,ift = i_dtft(ftX)
    overlayPlot([[1,signal],[0,iftX]])
```

### signal manipulation and combination

```python
def multiply(A,B):
    C = []
    for x,y in zip(A,B):
        C.append([x[0],x[1]*y[1]])
    return C
```

### F(f)*F(g) => F(f convolution g)
### multiplication of signal results in convolution of their fourier transform
## Example

```python
def multiplyX(A,B):
    X = []
    c = []
    for x,y in zip(A,B):
        a = [x['r'],x['i']]
        b = [y['r'],y['i']]
        z = mult(a,b)
        freq = x['f']
        amp = m.sqrt(z[0] * z[0] + z[1] * z[1])
        phase = m.atan2(z[1], z[0])
        X.append({ 'r': z[0], 'i': z[1], 'f': freq, 'amp': amp, 'phase': phase })
        c.append([freq,magnitude(z[0],z[1])])
    return X,c
```

```python
ftX1,ft1 = dtft(signal)
plotSignal(ft1)
ftX2,ft2 = dtft(H)
plotSignal(ft2)
ftX,ft = multiplyX(ftX1,ftX2)
plotSignal(ft)
```

```python
iftX,ift = i_dtft(ftX)
```

```python
plotSignal(signal)
plotSignal(H)
plotInverse(iftX)  # convolution of the above two
```

```python
def convolution(A,B):
    C = []
    ftX1,ft1 = dtft(A)
    ftX2,ft2 = dtft(B)
    ftX,ft = multiplyX(ftX1,ftX2)
    iftX,ift = i_dtft(ftX)
    for x in iftX:
        C.append([x['f'],x['r']])
    return C
```

```python
def analyze(x,y):      # plot signals, their multiplication and their convolution
    print('SIGNALS:')
    plotSignal(x)
    plotSignal(y)
    print('MULTIPLICATION:')
    plotSignal(multiply(x,y))
    print('CONVOLUTION:')
    plotSignal(convolution(x,y))
```

```python
analyze(signal,H)
```

```python
analyze(multsin,H)
```

```python
def winder(signal,freq):
    N = len(signal)
    signalPlot = [[],[]]
    average = []
    total = [0,0]
    for n in range(N):
        phi = (2*m.pi*freq*n)/N
        c = [m.cos(phi),-m.sin(phi)]
        temp = mult([signal[n][1],0],c)
        signalPlot[0].append(temp[0])
        signalPlot[1].append(temp[1])
        total[0] +=temp[0]
        total[1] +=temp[1]
    average = [total[0]/N,total[1]/N]
    return signalPlot,average
```

```python
def W(n,N):
    return [m.cos(2*m.pi*n/N),m.sin(2*m.pi*n/N)]
```

```python
def FFT(signal):  # the passed signal should be complex [x,y]
    if len(signal)==2:
        temp = signal[1] #mult(W(0,len(signal)),signal[1])
        return [add(signal[0],temp),sub(signal[0],temp)]
    else:
        even = [signal[i] for i in range(0,len(signal),2)]
        odd = [signal[i] for i in range(1,len(signal),2)]
        e,o = FFT(even),FFT(odd)
        transform = []
        o = [mult(W(-i,len(signal)),o[i]) for i in [*range(len(o))]]
        for i in range(len(signal)//2):
            transform.append(add(e[i],o[i]))
        for i in range(len(signal)//2):
            transform.append(sub(e[i],o[i]))
        return transform
```

```python
def FFT_fy(signal):
    try:
        s_time = [signal[i]['f'] for i in range(len(signal))]
        s_component = [[signal[i]['r'],signal[i]['i']] for i in range(len(signal))]
    except:  
        s_time = [signal[i][0] for i in range(len(signal))]
        try:
            len(signal[0][1])
            s_component = [signal[i][1] for i in range(len(signal))]
        except:
            s_component = [[signal[i][1],0] for i in range(len(signal))]
    s_fft = FFT(s_component)
    s_fft = [[s_fft[i][0]/len(s_fft),s_fft[i][1]/len(s_fft)] for i in range(len(s_fft))]
    s_fftmag = [magnitude(s_fft[i][0],s_fft[i][1]) for i in range(len(s_fft))]
    X = [{'r':s_fft[i][0],'i':s_fft[i][1],'f':s_time[i]} for i in range(len(s_time))]
    Comp = [[s_time[i],s_fftmag[i]] for i in range(len(s_time))]
    low_pass(X,Comp,len(signal)//2)
    return X,Comp
```

```python
def IFFT(signal):  # the passed signal should be complex [x,y]
    if len(signal)==2:
        temp = signal[1] #mult(W(0,len(signal)),signal[1])
        return [add(signal[0],temp),sub(signal[0],temp)]
    else:
        even = [signal[i] for i in range(0,len(signal),2)]
        odd = [signal[i] for i in range(1,len(signal),2)]
        e,o = FFT(even),FFT(odd)
        transform = []
        o = [mult(W(-i,len(signal)),o[i]) for i in [*range(len(o))]]
        for i in range(len(signal)//2):
            transform.append(add(e[i],o[i]))
        for i in range(len(signal)//2):
            transform.append(sub(e[i],o[i]))
        return transform
```

```python
def IFFT_fy(signal):
    try:
        s_time = [signal[i]['f'] for i in range(len(signal))]
        s_component = [[signal[i]['r'],signal[i]['i']] for i in range(len(signal))]
    except:  
        s_time = [signal[i][0] for i in range(len(signal))]
        try:
            len(signal[0][1])
            s_component = [signal[i][1] for i in range(len(signal))]
        except:
            s_component = [[signal[i][1],0] for i in range(len(signal))]
    s_fft = IFFT(s_component)
    s_fft.append(s_fft.pop(0))
    s_fft = list(reversed(s_fft))
    s_fft = [[2*s_fft[i][0],2*s_fft[i][1]] for i in range(len(s_fft))]
    s_fftmag = [magnitude(s_fft[i][0],s_fft[i][1]) for i in range(len(s_fft))]
    X = [{'r':s_fft[i][0],'i':s_fft[i][1],'f':s_time[i]} for i in range(len(s_time))]
    Comp = [[s_time[i],s_fftmag[i]] for i in range(len(s_time))]
    return X,Comp
```

```python
def processSignalFFT(signal, reflect = 0):
    %matplotlib inline
    plotSignal(signal)
    Xft,ft = FFT_fy(signal)
#     plotSignal(ft,reflect)
    plotSignal(ft[:len(ft)//2],reflect)
    iXft,ift = IFFT_fy(Xft)
    plotInverse(iXft)
```
